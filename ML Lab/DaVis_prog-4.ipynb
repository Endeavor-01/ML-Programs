# SCATTER PLOT

```python


import matplotlib.pyplot as plt
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [8934, 14902, 3409, 25672, 12300, 2023, 6890]
plt.scatter(days, steps_walked, c ="blue")
plt.title('STEP COUNT',fontsize=20)
plt.ylabel('STEPS WALKED')
plt.xlabel('DAYS')
# To show the plot
plt.show()
```

# BAR GRAPH

```python
# BAR CHART
plt.bar(days, steps_walked, color ='maroon', width = 0.4)
plt.xlabel("DAYS")
plt.ylabel("STEPS WALKED")
plt.title("STEPS COUNT")
plt.show()
```

# LINE PLOT

```python
plt.plot(days, steps_walked)  # Plot the chart
plt.title("Step count | This week and last week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked")
plt.show()
```

# PIE-CHART

```python
# Create a dataframe with the data
import pandas as pd
data = {'days' :["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
'steps_walked': [8934, 14902, 3409, 25672, 12300, 2023, 6890]}

df = pd.DataFrame(data)

# Plot the pie chart

plt.pie(df['steps_walked'], labels=df['days'])
plt.title('STEPS-COUNT')

plt.show()
```
