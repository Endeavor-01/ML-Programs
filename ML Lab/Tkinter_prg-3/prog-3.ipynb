```python
# READIG DATA USING TKINTER METHOD
# IN THIS METHOD NO NEED TO REMEMBER THE LOCATION OF THE DATASET FILE
from tkinter import Tk
from tkinter.filedialog import askopenfilename

Tk().withdraw()
filename = askopenfilename()
print(filename)

```
